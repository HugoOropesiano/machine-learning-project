## Predicciones sobre el conjunto de competición

En este cuaderno se utiliza el modelo final previamente entrenado y guardado para generar predicciones sobre el conjunto de datos de competición.

```python
import pandas as pd
import joblib
import pickle

#Cargar modelo final
modelo = joblib.load("modelo_final.joblib")

#Cargar datos de competición
with open('Competicion/bank_competition.pkl', 'rb') as f:
    df_comp = pickle.load(f)

#Transformacion de pdays
df_comp['contacted_before'] = (df_comp['pdays'] > -1).astype(int)
df_comp['pdays'] = df_comp['pdays'].apply(lambda x: 0 if x == -1 else x)

#Generar predicciones
predicciones = modelo.predict(df_comp)

#Convertir a etiquetas originales
le = joblib.load("label_encoder.pkl")
predicciones_labels = le.inverse_transform(predicciones)

#Guardar predicciones
df_predicciones = pd.DataFrame({'deposit': predicciones_labels})
df_predicciones.to_csv("predicciones.csv", index=False)

print("Predicciones guardadas")
df_predicciones.head()
```
